```python
import pandas as pd

df = pd.read_csv("../data/SuperStoreOrders.csv")

print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
total_sales = df["sales"].sum()

total_profit = df["profit"].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
top_products = df.groupby("product_name")["sales"].sum().sort_values(ascending=False)

print(top_products.head(10))
```

```python
region_sales = df.groupby("region")["sales"].sum()

print(region_sales)
```

```python
import matplotlib.pyplot as plt

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    format="mixed",
    dayfirst=True
)
```

```python
monthly_sales = df.groupby(df["order_date"].dt.to_period("M"))["sales"].sum()

print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.savefig("../outputs/monthly_sales_trend.png")

plt.show()
```

```python
category_sales = df.groupby("category")["sales"].sum().sort_values(ascending=False)

print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.savefig("../outputs/category_sales.png")

plt.show()
```

```python
category_profit = df.groupby("category")["profit"].sum().sort_values(ascending=False)

print(category_profit)
```

```python
category_profit.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.savefig("../outputs/category_profit.png")

plt.show()
```

```python
top_products = df.groupby("product_name")["sales"].sum().sort_values(ascending=False)

print(top_products.head(10))
```

```python
top_products.head(10).plot(kind="bar")

plt.title("Top 10 Products")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.xticks(rotation=75)

plt.savefig("../outputs/top_products.png")

plt.show()
```

# Business Insights

1. Technology and Furniture categories generated the highest revenue.

2. Certain regions consistently outperformed others in total sales.

3. Some products generated high sales volume but lower profitability.

4. Revenue trends showed seasonal fluctuations over time.

5. Discounts affected profit margins in several product categories.

# Recommendations

1. Focus marketing efforts on high-performing product categories.

2. Improve sales strategies in lower-performing regions.

3. Reduce excessive discounts on low-profit products.

4. Increase inventory for top-selling products.

5. Use sales trend analysis to plan seasonal promotions.
